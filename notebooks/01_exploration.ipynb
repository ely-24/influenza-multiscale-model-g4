A Influenza A é um dos vírus de evolução mais rápida entre os patógenos humanos, apresentando taxas de mutação muito superiores às dos vírus Influenza B e C. Esta elevada taxa de mutação, sobretudo nos genes que codificam as glicoproteínas de superfície hemaglutinina (HA) e neuraminidase (NA), provoca alterações contínuas nas regiões antigénicas — um processo conhecido como *antigenic drift*. Essas modificações permitem ao vírus escapar parcialmente à imunidade pré-existente da população, favorecendo o surgimento de novas variantes com capacidade aumentada de persistir e propagar-se. (https://www.sciencedirect.com/science/article/pii/S1567134823000229?via%3Dihub#bb0090) 

As mutações que afetam genes como **HA**, **NA**, **PB1**, **PA** e **NS1** podem alterar características fundamentais do vírus, como a afinidade pelo recetor, a estabilidade estrutural, a eficiência da polimerase e a evasão da resposta imunitária. Estas alterações modificam o **fitness viral**, isto é, a capacidade do vírus replicar-se e transmitir-se com sucesso. Variantes com maior fitness replicam-se mais rapidamente, alcançam cargas virais mais elevadas e transmitem-se com maior eficiência entre indivíduos, originando cadeias de transmissão mais longas e favorecendo a substituição de linhagens menos adaptadas. https://www.nature.com/articles/nature13087, https://www.nature.com/articles/ncomms8952, https://www.worldscientific.com/doi/10.1142/S0218339025500068

Ao nível populacional, as diferenças no fitness viral repercutem-se nas **curvas epidémicas**, influenciando a magnitude, a duração e o ritmo das epidemias sazonais. Alterações aparentemente pequenas ao nível molecular podem, assim, traduzir-se em grandes variações na dinâmica epidemiológica observada, moldando a incidência anual da gripe e a necessidade de atualização periódica das vacinas. https://elifesciences.org/articles/91849, https://journals.plos.org/plospathogens/article?id=10.1371/journal.ppat.1007291

## Introdução

A Influenza A é um dos maiores desafios de saúde pública no mundo. Todos os anos, causa epidemias sazonais que atingem milhões de pessoas e têm um impacto considerável na mortalidade e na sobrecarga dos sistemas de saúde. Trata-se de uma infeção respiratória aguda, com circulação global, que se transmite facilmente entre indivíduos. Estima-se que, anualmente, ocorram cerca de mil milhões de casos, dos quais entre 3 e 5 milhões são graves e até 650 mil acabam por ser fatais. Os grupos mais afetados incluem crianças pequenas, idosos, grávidas, pessoas com doenças crónicas ou sistemas imunitários debilitados, além dos profissionais de saúde, que convivem diariamente com o risco de exposição (World Health Organization, 2023).

A gripe sazonal tem um caráter previsível em regiões temperadas, surgindo geralmente nos meses frios, mas em zonas tropicais pode ocorrer ao longo de todo o ano, em surtos menos regulares. Esta natureza cíclica relaciona-se com fatores climáticos e comportamentais, mas também com a notável capacidade de mutação do vírus Influenza A. As suas estirpes mais comuns em humanos, A(H1N1)pdm09 e A(H3N2), passam por alterações genéticas constantes que lhes permitem ultrapassar a imunidade adquirida e desafiar a eficácia das vacinas. Por isso, a vigilância contínua desempenha um papel essencial na prevenção e controlo da doença.

A Organização Mundial da Saúde (OMS), através do Global Influenza Surveillance and Response System (GISRS), acompanha a atividade do vírus em todo o mundo e, duas vezes por ano, recomenda as composições mais adequadas para as vacinas sazonais. Além disso, orienta campanhas de vacinação, coordena respostas a surtos e apoia os países no reforço das suas capacidades laboratoriais e de monitorização. Esta rede internacional é vital para identificar novas variantes e responder rapidamente a ameaças emergentes.

Entre os vírus respiratórios humanos, a Influenza A destaca-se pela rapidez com que evolui. A sua taxa de mutação é muito mais elevada que a dos vírus Influenza B ou C, especialmente nos genes que codificam as proteínas de superfície — hemaglutinina (HA) e neuraminidase (NA). Estas proteínas estão na linha da frente do reconhecimento pelo sistema imunitário, e mesmo pequenas alterações nas suas estruturas — conhecidas como antigenic drift — podem permitir ao vírus escapar às defesas geradas por infeções ou vacinas anteriores (Krammer et al., 2018).

Os efeitos dessas mutações vão muito além da escala molecular. Mudanças em genes como HA, NA, PB1, PA e NS1 influenciam diretamente a forma como o vírus se comporta: alteram a afinidade pelos recetores, a estabilidade da partícula viral, a precisão da sua replicação e a forma como contorna a resposta imunitária. Essas variações determinam o fitness viral, ou seja, o “sucesso biológico” do vírus em se multiplicar e espalhar. Cepas com maior fitness tendem a replicar-se mais rapidamente, gerar cargas virais mais altas e transmitir-se com maior eficiência, prolongando as cadeias de transmissão e, em muitos casos, substituindo linhagens menos adaptadas (Nobusawa & Sato, 2014; McWhite et al., 2015; Chen et al., 2025).

Quando observadas em escala populacional, essas pequenas mudanças genéticas refletem-se nas curvas epidémicas — influenciando o tamanho, a duração e a intensidade das epidemias. Pequenas alterações no genoma do vírus podem, portanto, moldar a dinâmica global das gripes sazonais e até determinar a necessidade de atualização das vacinas (Wen et al., 2024; Koelle & Rasmussen, 2019).

Apesar dos avanços no estudo da evolução viral, ainda é difícil estabelecer uma relação direta entre mutações específicas e parâmetros epidémicos — como o número de reprodução básica (R0), o alcance e o formato das epidemias. Essa dificuldade decorre da complexidade dos processos que ocorrem dentro do hospedeiro, das coinfeções em células, das interações entre mutações (epistasia) e da grande diversidade imunitária entre indivíduos. Unir estas escalas — do nível molecular ao populacional — continua a ser um dos maiores desafios da ciência das doenças infeciosas (Yan et al., 2021; Koelle & Rasmussen, 2019).



### Definição do Problema

O problema específico abordado neste trabalho pode ser formulado da seguinte forma:

**Podem mutações na proteína Hemagglutinina (HA) do vírus Influenza A ser utilizadas para estimar um score de aptidão associado à transmissibilidade e, de que forma, esse score influencia a propagação epidémica do vírus a nível populacional?**

Ao responder a esta questão, o estudo procura avaliar o potencial de modelos computacionais baseados em dados moleculares como ferramentas de apoio à vigilância epidemiológica e à tomada de decisão em saúde pública.


### Objetivos

O presente trabalho tem como objetivo geral desenvolver um modelo computacional multi-escala que relacione alterações moleculares no vírus Influenza A com o seu impacto na dinâmica de disseminação populacional.

De forma específica, este estudo propõe-se a:

- **Analisar mutações na proteína Hemaglutinina (HA) do vírus Influenza A**, utilizando dados de sequenciação genómica provenientes de bases de dados públicas;
- **Inferir um proxy de aptidão viral (fitness) associado à transmissibilidade**, recorrendo a técnicas de *machine learning* aplicadas a características moleculares das sequências de HA;
- **Integrar o fitness previsto num modelo epidemiológico baseado em agentes (Agent-Based Model, ABM)**, de modo a simular a disseminação de diferentes variantes virais a nível populacional;
- **Validar empiricamente o comportamento do modelo**, comparando os resultados das simulações com dados epidemiológicos reais de vigilância da gripe, nomeadamente provenientes da plataforma FluNet da Organização Mundial da Saúde.

Esta abordagem permite estabelecer uma ligação mecanística entre o nível molecular e o nível populacional, contribuindo para uma compreensão mais integrada da evolução viral e do seu impacto epidemiológico.
